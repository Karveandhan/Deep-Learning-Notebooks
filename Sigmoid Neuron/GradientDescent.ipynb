```python
import numpy as np
```

```python
X=[0.5,0.2]
```

```python
Y=[2.5,0.9]
```

```python
def f(w,b,x):
    return 1.0/(1.0+np.exp(-(w*x+b)))
```

```python
def error(w,b):
    err=0.0
    for x,y in zip(X,Y):
        fx=f(w,b,x)
        err=err+np.pow((fx-y),2)
    return err
```

```python
def grad_b(w,b,x,y):
    fx=f(w,b,x)
    return (fx-y)*fx*(1-fx)
```

```python
def grad_w(w,b,x,y):
    fx=f(w,b,x)
    return (fx-y)*fx*(1-fx)*x
```

```python
def gradient_descent():
    w,b,eta=-2,-2,1.0
    max_epochs=1000
    for i in range(max_epochs):
        dw,db=0,-8
        for x,y in zip(X,Y):
            dw=dw+grad_w(w,b,x,y)
            db=db+grad_b(w,b,x,y)
        w=w-eta*dw
        b=b-eta*db
        print(w,b)
```
